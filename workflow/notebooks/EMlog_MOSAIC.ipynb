```python
# drift
import numpy as np
```

```python
N = 200 # number of haploids
f = 0.08 # starting freq
T = 20 # number of generations
S = 4000 # number of sites

def initialize(N, f):
    g0 = np.zeros(N, dtype = 'short')
    g0[0:int(N*f)] =1
    return(g0)


def reproduce(g0):
    return(np.random.choice(a = g0, size=len(g0), replace=True))

def evolve(g0, T):
    af = np.zeros(T+1, dtype = 'short')
    af[0] = g0.sum()
    g = reproduce(g0)
    af[1] = g.sum()
    for i in range(2, T+1):
        g = reproduce(g)
        af[i] = g.sum()
    return(af)
```

```python
g0 = initialize(N, f)
res = np.zeros((S, T+1), dtype = 'short')
for i in range(S):
    res[i] = evolve(g0, T)
```

```python
binned = [np.bincount(res[:, i], minlength = N) for i in range(T+1) ]
```
